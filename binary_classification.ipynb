```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (7.0, 7.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'bone'
```

```python
def f(X):
    y = []
    for i in range(len(X)):
        if (X[i][0])**2 + (X[i][1])**2 < 0.8 and (X[i][0])**2 + (X[i][1])**2 > 0.4:
            y.append(1)
        elif abs(X[i][0]) < 0.1 or abs(X[i][1]) < 0.1:
            y.append(1) 
        else:
            y.append(0)
    return np.array(y)

X = np.random.rand(1000, 2) * 2.0 - 1.0
y = f(X)
```

```python
f = open("data.ssv", "w")

f.write("{ln}\n".format(ln = len(X)))

for i in range(len(X)):
    txt = "{x1} {x2} {y}\n".format(x1 = X[i][0], x2 = X[i][1], y = y[i])
    f.write(txt)
    
f.close()
```

```python
plt.scatter(X[:,0], X[:,1], c = y.astype(int), cmap='bone')
```

```python
f = open("result.ssv", "r")

X1 = []
X2 = []
Y = []

while True:
    data = list(map(float, f.readline().split()))
    if(len(data) == 0):
        break
        
    X1.append(data[0])
    X2.append(data[1])
    Y.append(data[2])


X1 = np.array(X2[:200])
X2 = np.array(X2[:200])
Y = np.array(Y).reshape((200, 200)).T

f.close()
```

```python
plt.contourf(X1, X2, Y, cmap='bone')
```
